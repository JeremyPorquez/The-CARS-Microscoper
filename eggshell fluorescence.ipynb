Image subtraction

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
im1 = ndimage.imread("eggshell_zm3_hor001.tif")
im2 = ndimage.imread("eggshell_zm3_vert001.tif")
```

```python
plt.rcParams['figure.figsize'] = (15.0, 5.0)
fig = plt.figure()
plt.subplot(131)
plt.imshow((im1),cmap='gray')
plt.subplot(132)
plt.imshow((im2),cmap='gray')
plt.subplot(133)
plt.imshow((im1-im2),cmap='gray',vmax=50)
```

```python
plt.imshow(im1*np.abs(im1-im2),cmap='gray',vmin=1)
```
