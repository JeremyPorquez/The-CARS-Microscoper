```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-10,10,0.01)
def y(x,z):
    return (np.sin(x*z)/x)**2
```

```python
f = plt.figure()
plt.axis([-10,10,0,0.6])
for i in np.arange(0,1,0.25):
    plt.plot(x,y(x,i),lw=5)
plt.xlabel(r'$\Delta k$', fontsize=30)
plt.ylabel(r'$sin(\Delta k z)/z$', fontsize=30)
plt.show()
```
