#The susceptibility $\chi^{(3)}$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def chi(delta):
    return 1/(-delta - 1j*G)
G = 1
d = np.arange(-10,10,0.1)
plt.figure(figsize=(8,6))
plt.title(r'\chi^{(3)}')
plt.plot(d,chi(d), label='dispersive (real)')
plt.plot(d,chi(d)*np.conj(chi(d)), label='resonant (imaginary)')
plt.legend(loc='upper right')
```
